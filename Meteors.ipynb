# Introduction: Characterising Meteor Showers

Meteor showers are captivating celestial events that have fascinated astronomers and stargazers for centuries. These showers occur when the Earth passes through debris, 
typically left behind by comets, causing a series of bright streaks in the night sky as the particles burn up in the Earth's atmosphere.

### *What Do We Know About the Phenomenon*

The understanding of meteor showers has progressed throughout the last century, revealing they result from Earth passing through debris trails left by comets or sometimes asteroids. As these particles enter the atmosphere at high speeds, they disintegrate, creating visible meteors.
The article "**Meteor Phenomena and Bodies**" comprehensively explores the mechanisms, occurrences, and celestial bodies involved in these events. 

#### "Meteor Phenomena and Bodies"

The above mentioned Review Paper informs us in short about the following: Meteoroids entering Earth's atmosphere being observed as meteors, and many methods we use to put some numbers and  science on the phenomena. These methods provide detailed information on their atmospheric path, behavior, and physical changes as they burn up, along with the radiation they emit and their ionizing effects. 
Researchers also study the sounds they create, and use theoretical models to understand their interactions with the atmosphere. Observational data is enhanced by radar and spectroscopic methods, as well as studies of artificial meteors. Understanding meteoroids' composition and orbit connects them to larger celestial bodies like asteroids and comets. Observations of meteor showers and storms help assess the influx rate of meteoroids to Earth and potential related hazards.

### *Open questions*

Despite our knowledge, several open questions persist in the field of meteor science. These include the topics of :

  * **Composition**
  * **Orbital Path Prediction**
  * **Parent Bodies**
  * **Interaction with the Atmosphere**

### *What are we actually looking at?*

When we observe a meteor shower, we are essentially looking at the interaction between meteoroid particles (typically composed of dust, rock, and ice) and the Earth's atmosphere. 
These particles, traveling at speeds of up to 71 km/s, undergo rapid ionization and vaporization due to atmospheric friction, creating the characteristic bright streaks in the sky.

## *Original Idea*

One of the first significant papers, *"Interplanetary Debris Near the Earth"*, that looked into the characterization of meteor showers was written by Gerald S. Hawkins in the 20th century. ## *Characterizing the daytime sextantids meteor shower*

Lastly we are looking at a paper from 2022 named ‘Characterizing the daytime sextantids meteor shower and unveiling the nature of the phaethon-geminid stream complex’, which draws parallels between the Daytime Sextantids (also referred to as DSX) and the asteroid 2005 UD. Furthermore this research compared the DSX to the Geminids (another prominent meteor shower) and found that they possibly share a common origin.

The mass of the DSX stream is estimated to be 10^16 g, the same as asteroid 2005 UD, suggesting the stream is too massive to have been created by recent meteoroid production from 2005 UD. However the paper proposes that the DSX and 2005 UD share a common origin among the break-up event that created 3200 Phaethon.

## Type of distribution
The Figure below depicts the distribution of absolute magnitude of Meteors from EDMOND Database (2016-Dataset) and we can conclude that the distribtuion resembels a normal (Gaussian) distribution. The mean is on the right hand side of the median, thus the skewness of the distribution is negative.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data from CSV
df = pd.read_csv('EDM.csv')  # Replace 'your_file.csv' with your file name

# Ensure the '_amag' column is numeric
df['_amag'] = pd.to_numeric(df['_amag'], errors='coerce')

# Calculate mean and median
mean_value = df['_amag'].mean()
median_value = df['_amag'].median()

# Plotting the histogram
plt.hist(df['_amag'].dropna(), bins=40, edgecolor='black')  # You can adjust the number of bins
plt.xlabel('Absolute Magnitude (mag)')
plt.ylabel('Number of Meteors')
plt.title('Brightness of Meteors')

# Plot mean and median lines
plt.axvline(mean_value, color='red', linestyle='dashed', linewidth=2, label=f'Mean: {mean_value:.2f}')
plt.axvline(median_value, color='limegreen', linestyle='dashed', linewidth=2, label=f'Median: {median_value:.2f}')

# Adding a legend
plt.legend()

# Add grid and show plot
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load data
data = pd.read_csv('EDM.csv')  # Replace with your file path
terminal_heights = data['_H1']

# Calculate statistics
mean = terminal_heights.mean()
std_dev = terminal_heights.std()
variance = std_dev ** 2

# Create histogram
plt.hist(terminal_heights, bins=50, edgecolor='black', alpha=0.7)

# Add mean, standard deviation, and variance lines
plt.axvline(mean, color='red', linestyle='dashed', linewidth=1.7, label=f'Mean: {mean:.2f}')
plt.axvline(mean - std_dev, color='green', linestyle='dashed', linewidth=1, label=f'Standard Deviation: {std_dev:.2f}')
plt.axvline(mean + std_dev, color='green', linestyle='dashed', linewidth=1)
plt.axvline(mean - np.sqrt(variance), color='yellow', linestyle='dashed', linewidth=1.7, label=f'Variance: {variance:.2f}')
plt.axvline(mean + np.sqrt(variance), color='yellow', linestyle='dashed', linewidth=1.7)

# Add labels and title
plt.xlabel('Terminal Height (km)')
plt.ylabel('Number of Meteors')
plt.title('Terminal Height of Meteors')
plt.legend()
plt.grid()

# Show plot
plt.show()
```

## Probabilities
Here we have a simple probability calculation to show the chance of Sporadic meteors and other streams.

```python
import pandas as pd

df = pd.read_csv('EDM.csv')

# Calculate the probabilities
total_entries = len(df)
spo_count = df['_stream'].value_counts().get('_spo', 0)
spo_probability = spo_count / total_entries
other_probability = 1 - spo_probability

print("Probability of Sporadic meteors:", spo_probability)
print("Probability of other Streams:", other_probability)
```

As we can see the probability of Sporadic meteors is very high, we will show with a graph what that actually means.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
df = pd.read_csv('EDM.csv')  # Replace 'your_file.csv' with your file path

# Filter the data based on magnitude
df = df[(df['_amag'] >= -4) & (df['_amag'] <= 2)]

# Define the groups
groups = {
    '_J8_LYR': 'red',
    '_J8_ELY': 'red',
    '_J8_STA': 'blue',
    '_J8_COM': 'purple',
    '_J8_QUA': 'yellow',
    '_J8_GEM': 'green',
    '_J8_DRA': 'brown',
    '_J8_PER': 'orange'
}

# Plotting
plt.figure(figsize=(10, 6))

for stream, color in groups.items():
    # Filter data for each stream
    stream_data = df[df['_stream'] == stream]
    plt.scatter(stream_data['_ra_o'], stream_data['_dc_o'], c=color, marker='*', label=stream)


colors = ['orange', 'red', 'blue', 'purple', 'yellow', 'green', 'brown']
labels = ['Perseids', 'Lyrids', 'Southern Taurids', 'Comae Berenicids', 'Quandrantids', 'Geminids', 'October Draconids',]
from matplotlib.lines import Line2D
handles = [Line2D([0], [0], color=color, marker='*', linestyle='', markersize=10) for color in colors]



plt.xlabel('Right Ascension (°)')
plt.ylabel('Declination (°)')
plt.title('Scatter Plot of Right Ascension vs Declination by Stream')
plt.grid(True)
plt.legend(handles, labels)
plt.show()

# 2nd graph
df = pd.read_csv('EDM.csv')  # Replace 'your_file.csv' with your file path

# Filter the data based on magnitude
df = df[(df['_amag'] >= -1) & (df['_amag'] <= 0)]

# Define the groups
groups = {
    '_J8_LYR': 'red',
    '_J8_ELY': 'red',
    '_J8_STA': 'blue',
    '_J8_COM': 'purple',
    '_J8_QUA': 'yellow',
    '_J8_GEM': 'green',
    '_J8_DRA': 'brown',
    '_J8_PER': 'orange',
    '_spo' : 'peachpuff'
}

# Plotting
plt.figure(figsize=(10, 6))

for stream, color in groups.items():
    if stream != '_spo':  # Skip the emphasized group for now
        stream_data = df[df['_stream'] == stream]
        plt.scatter(stream_data['_ra_o'], stream_data['_dc_o'], c=color, label=stream, marker='*')

# Now plot the emphasized group
stream_data = df[df['_stream'] == '_spo']
plt.scatter(stream_data['_ra_o'], stream_data['_dc_o'], c=groups['_spo'], label='_spo', marker='.', s=10, edgecolor='peachpuff')


colors = ['orange', 'red', 'blue', 'purple', 'yellow', 'green', 'brown', 'peachpuff' ]
labels = ['Perseids', 'Lyrids', 'Southern Taurids', 'Comae Berenicids', 'Quandrantids', 'Geminids', 'October Draconids', 'Sporadic Meteors', ]
from matplotlib.lines import Line2D
handles = [Line2D([0], [0], color=color, marker='*', linestyle='', markersize=10) for color in colors]


plt.xlabel('Right Ascension (°)')
plt.ylabel('Declination (°)')
plt.title('Scatter Plot of Right Ascension vs Declination by Stream')
plt.grid(True)
plt.legend(handles, labels, loc='lower right')
plt.show()

```
